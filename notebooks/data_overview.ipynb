Select the UEFA Women's Euro Competition ID and select all of the matches and get the list of match IDs

```python
women_euro_id = women_euro.iloc[0]['competition_id']
women_euro_season = women_euro.iloc[0]['season_id']
print(f'Competition ID: {women_euro_id}')
print(f'Season ID: {women_euro_season}')
```
